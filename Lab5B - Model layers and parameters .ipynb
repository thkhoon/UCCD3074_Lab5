## 1. Build a network

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv_block1 = nn.Sequential(
            nn.Conv2d(3, 5, 2, stride=2), #in_channels, #out_channels, #kernel_size, stride
            nn.ReLU()
        )
        self.fc1 = nn.Linear(1280, 3)

    def forward(self, x):
        x = self.conv_block1(x)
        x = x.view(x.size(0), -1)
        x = self.fc1(x)
        return x
```
